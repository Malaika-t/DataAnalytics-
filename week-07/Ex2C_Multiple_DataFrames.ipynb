# Exercise 2.C Working with multiple DataFrames

```python
import numpy as np
import pandas as pd
```

## Basic Aggregations

```python
data = {'Product': ['A', 'B', 'C', 'A', 'B', 'C', 'A', 'B', 'C'], 
        'Region': ['North', 'South', 'East', 'West', 'North', 'South', 'East', 'West', 'North'], 
        'Units_Sold': [10, 15, 20, 5, 10, 15, 10, 5, 15],
        'Price': [100, 200, 150, 100, 200, 150, 100, 200, 150]}

df_sales = pd.DataFrame(data)
df_sales
```

```python
df_sales['Revenue'] = df_sales['Units_Sold'] * df_sales['Price']
df_sales
```

```python
total_units = df_sales['Units_Sold'].sum()
total_revenue = df_sales['Revenue'].sum() 

print(f'Total Units Sold: {total_units}')
print(f'Total Revenue: {total_revenue}')
```

```python
avg_units = df_sales['Units_Sold'].mean()
avg_revenue = df_sales['Revenue'].mean()

print(f'Average Units Sold: {avg_units:.2f}')
print(f'Average Revenue: {avg_revenue:.2f}')
```

## Grouping Data

```python
by_product =df_sales.groupby('Product')[['Units_Sold', 'Revenue']].sum()
print('Total by Product:')
by_product
```

```python
by_region = df_sales.groupby('Region')[['Units_Sold', 'Revenue']].mean()
print('Average by Region:')
by_region
```

```python
by_product_region = df_sales.groupby(['Product', 'Region'])[['Units_Sold', 'Revenue']].sum()
print('Total by Product and Region:')
by_product_region
```

## Sorting and Ranking Data

```python
df_sales_sorted_units = df_sales.sort_values('Units_Sold', ascending=True)
print('Sorted by Units Sold (ascending):')
df_sales_sorted_units
```

```python
df_sales_sorted_revenue = df_sales.sort_values('Revenue', ascending=False)
print('Sorted by Revenue (desecending):')
df_sales_sorted_revenue
```

```python
df_sales['Revenue_Rank'] = df_sales['Revenue'].rank(ascending=False).astype(int)
df_sales
```

## Merging DataFrames

```python
product_data = {'Product': ['A', 'B', 'C'], 
                'Category': ['Electronics', 'Furniture', 'Appliances']}
df_products = pd.DataFrame(product_data)
df_products
```

```python
df_merged = pd.merge(df_sales, df_products, on='Product')
df_merged 
```

```python
revenue_by_category = df_merged.groupby('Category')['Revenue'].sum()
print('Total Revenue by Category:')
revenue_by_category
```
