# Exercise 3.A Getting familiar with Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Creating Basic Plots

```python
x = [1, 2, 3, 4, 5] 
y = [2, 3, 5, 7, 11]
```

```python
plt.plot(x, y)
plt.show
```

```python
plt.plot(x, y, marker='o')
plt.show
```

```python
plt.plot(x, y, marker='o')
plt.title('Example Graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show
```

## Styling Plots

```python
plt.plot(x, y, 
         linestyle='--', 
         color='red',
         marker='s',
         markersize=10,
         markerfacecolor='green'
        )
plt.title('Example Graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show
```

```python
plt.plot(x, y,
         linestyle=':',
         color='blue',
         marker='D',
         markersize=8,
         markerfacecolor='orange'
)
plt.title('Example Graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show
```

```python
plt.plot(x, y, 
         linestyle='-.',
         color='purple',
         marker='^',
         markersize=12,
         markerfacecolor='yellow'
        )
plt.title('Example Graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show
```
