# Exercise 1.A NumPy: Manipulating Arrays

```python
import numpy as np
```

## Looping Through Arrays

```python
arrA = np.array([
    [11, 12, 13],
    [14, 15, 16]
])
print(arrA)
```

```python
for x in arrA:
    print(x * 2)
```

```python
for x in arrA:
    for y in x:
        print(y * 2)
```

**Difference between the first loop and the second loop:**

The first loop(`for x in arrA`) iterates over the rows of the 2D array. `x` is an entire row, so multiplying by 2 doubles every element in that row at once and prints the whole row.

The second loop(`for y in x`) iterates over each individual element within each row. So instead of printing a whole row at a time, it prints each single number multiplied by 2. The second loop gives us 6 individual printed values, while the first gives us 2 printed rows.

```python
for x in arrA:
    for y in x:
        loop_result = x * 2
        print(loop_result)

# loop_result = x * 2 uses x (the row), NOT y (the individual element).
# Even though the inner loop iterates over individual elements (y)
# x * 2 dpubles the entire row, so loop_result is a 1D array like [22, 24, 26], not a single number.
# If loop_result = y * 2 is used, we would get single numbers instead.
```

## Joining and Stacking Arrays

```python
joined_array_1 = np.concatenate((arrA, arrA))
print(joined_array_1)
```

```python
joined_array_2 = np.concatenate((arrA, arrA), axis=1)
print(joined_array_2)
```

```python
stacked_array_1 = np.stack((joined_array_1, joined_array_1))
print(stacked_array_1)
```

```python
stacked_array_2 = np.stack((joined_array_1, joined_array_2))
print(stacked_array_2)

# It gives ValueError:all input arrays must have the same shape.
# np.stack() requires all input arrays to be of the same shape.
# joined_array_1 is shape (4, 3) and joined_array_2 has shape (2, 6)
# So NumPy raises a ValueError because arrays of different shapes can't be stacked.
```

```python
another_joined_array = np.concatenate((joined_array_1, arrA))
print(another_joined_array)
```

```python
another_2 = np.concatenate((joined_array_1, arrA), axis=1)
print(another_2)

# It gives ValueError: all the input array dimensions except for the concatenation axis must match exactly, but along dimension 0, the array at index 0 has size 4 and the array at index 1 has size 2
# This fails because concatenating on axis=1 (columns) requires both arrays to have the same number of ROWS. 
# joined_array_1 has 4 rows and arrA has 2 rows
# so NumPy cannot line them up side by side as the row counts don't match.
```

```python
h_stacked = np.hstack((arrA, arrA))
print("hstack result:")
print(h_stacked)

# hstack is horizontal stack
# It joins arrays side by side (along columns, axis=1)
```

```python
v_stacked = np.vstack((arrA, arrA))
print("vstack result:")
print(v_stacked)

# vstack is vertical stack 
# It joins arrays on top of each other (along rows, axis=0)
```

```python
d_stacked = np.dstack((arrA, arrA))
print("dstack result:")
print(d_stacked)

# dstack is depth stack 
# It joins arrays along the third axis (creates 3D array)
```

## Splitting Arrays

```python
splitsky = np.array_split(joined_array_1, 2)
print(splitsky)

# splitsky contains 2 arrays
# Each of the 2 resulting arrays has 2 dimensions.
```

```python
spliterino = np.array_split(joined_array_1, 3)
print(spliterino)

# spliterino contains 3 arrays
# joined_array_1 has 4 rows; splitting into 3 gives arrays of sizes 2, 1, and 1.
# Each of the 3 resulting arrays is still 2-dimensional.
```

## Searching and Replacing Array Values

```python
result_1 = np.where(arrA % 2 == 0, arrA, 'odd')
print(result_1)
```

## Sorting and Filtering Arrays,

```python
arrB = np.array([[7, 6], [13, 5], [7, 3]])

arrB_sorted = np.sort(arrB)
print(arrB_sorted)
```

```python
print(arrB)

z = arrB < 7

arrC = arrB[z]
print(arrC)
```

```python
y = arrB >= 7
arrD = arrB[y]
print(arrD)
```
