# Exercise 3.B More plotting practice

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Advanced Plotting Techniques

```python
products = ['Product A', 'Product B', 'Product C', 'Product D'] 
sales = [23, 17, 35, 29]

plt.bar(products, sales)
plt.show()
```

```python
plt.bar(products, sales, color='skyblue', width=0.5)
plt.title('Sales by Product')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.show()
```

```python
x = np.random.rand(50) 
y = np.random.rand(50)

plt.scatter(x, y)
plt.show()
```

```python
x = np.random.rand(50) 
y = np.random.rand(50)
s =np.random.rand(50) * 100

plt.scatter(x, y, s=s, c=s, cmap='viridis')
plt.colorbar(label='Marker Size')
plt.title('Scatter Plot - Variable Marker Size')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

## Subplots

```python
fig, (ax1) = plt.subplots(1, figsize=(12, 5))

# Subplot 1: Sine wave (Line Chart)
x = np.linspace(0, 10, 100)
y = np.sin(x)
ax1.plot(x, y, color='blue')
ax1.set_title('Sine Wave')
ax1.set_xlabel('X')
ax1.set_ylabel('sine(x)')
```

```python
fig, (ax2) = plt.subplots(1, figsize=(12, 5))

# Subplot 2: Histogram
data = np.random.randn(1000)
ax2.hist(data, bins=30, color='purple', edgecolor='black')
ax2.set_title('Histogram of Random Numbers')
ax2.set_xlabel('Value')
ax2.set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```
