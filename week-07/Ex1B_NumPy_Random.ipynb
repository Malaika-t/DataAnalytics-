## Generating Random Choices from an Array

```python
students = np.array([
    "Alice", "Bob", "Charlie", "Dan", "Ethan", "Julia", "Kevin", "Hannah", "Sarah", "John", "Noah", "Ivan"
])
```

```python
call_on = np.random.choice(students)
print(call_on)
```

```python
breakout_groups = np.random.choice(students, size=(3, 4), replace=False)
print(breakout_groups)

# This randomly assigns 12 students into 3 groups of 4.
# It works well as there are exactly 12 students and need 12 spots (3x4).
# replace=False ensures that no student appears twice. 
# This is an effective way to assign breakout groups because its fair, random, and unbiased.
```

```python
probabilities = [0.07, 0.3, 0.07, 0.07, 0.07, 0.07, 0.0, 0.07, 0.07, 0.07, 0.07, 0.07]
call_on_weighted = np.random.choice(students, size=5, p=probabilities)
print(call_on_weighted)

# Probabilities values assigned:
# - Kevin: 0% chance (0.0)
# - Bob: 30% chance (0.3)
# - Remaining 10 students share the other 70% evenly that is 0.07 each 
```
