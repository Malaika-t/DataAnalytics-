## Sample Arrays

```python
r_1 = np.random.randint(1, 21, size=2)
r_2 = np.random.randint(1, 21, size=(1, 2))
r_3 = np.random.random(size=(2,1))
r_4 = np.random.randint(20, 101, size=(4, 5))
r_5 = np.random.random(size=(4, 5))
```

## Permutations

```python
print(r_4)
```

```python
np.random.shuffle(r_4)
print(r_4)
```

Shuffle() rearranges only the ROWS of a 2D array, not individual elements.
The elements within each row stay in their original order; only the row order changes.

```python
new_shuffle = np.random.permutation(r_4)
print(new_shuffle)
print(r_4)

# Permutation() returns a new shuffled copy but r_4 is unchanged.
# Shuffle() modifies r_4 in place and returns None.
```

Use shuffle when you want to reorder an array in place and don't need to keep the original.
Use premutation when you need a shuffled copy and want to preserve the original array.

## Mathematical Operations

```python
print(r_1, r_1.shape)
print(r_2, r_2.shape)
result = np.add(r_1, r_2)
print(result, result.shape)
```

```python
print(r_1, r_1.shape)
print(r_3, r_3.shape)
result2 = np.add(r_1, r_3)
print(result2, result2.shape)
```

```python
print(r_2, r_2.shape)
print(r_4, r_4.shape)
result3 = np.add(r_2, r_4)
print(result3, result3.shape)

# ValueError: r_2 shape (1, 2) + r_4 shape (4, 5) are not compatible.
```

```python
try: 
    print(np.add(r_1, r_4))
except ValueError as e:
    print("r_1 + r_4 error:", e)

try: 
    print(np.add(r_2, r_4))
except ValueError as e:
    print("r_2 + r_4 error:", e)

try: 
    print(np.add(r_3, r_4))
except ValueError as e:
    print("r_3 + r_4 error:", e)

# None of these are compatible with r_4 shape(4, 5)
```

```python
print(np.add(np.array([1]), r_4))
# (1,) broadcasts to every element -> matches(4, 5)
```

```python
print(np.add(np.array([[1]]), r_4))
# (1, 1) broadcasts across rows and columns -> matches(4, 5)
```

```python
print(np.add(np.array([1,1,1,1,1]), r_4))
# (5,) broadcasts across each row -> matches(4, 5)
```

```python
print(np.add(np.array([[1],[1]]), r_4))
# ValueError: operands could not be broadcast together with shapes (2,1) (4,5)
```

Broadcasting rule: NumPy compares shapes from the Right. Dimensions are compatible if they are equal or one of them is 1. The shape with 1 gets stretched to match the other. The resulting array takes the larger size in each dimension. Incompatible shapes raise ValueError.

## Grocery Problem

```python
# [vegetables, fruits, snacks, beverages]
bag1 = np.array([2, 3, 4, 0])
bag2 = np.array([4, 0, 1, 2])
print(bag1)
print(bag2)
```

```python
total_bag1 = np.sum(bag1)
total_bag2 = np.sum(bag2)
print(f"The first bad contains {total_bag1} items, and the second bag contains {total_bag2} item.")
```

```python
totals = np.add(bag1, bag2)
print(f"There are {totals[0]} vegetables, {totals[1]} fruits, {totals[2]} boxes of snacks, and {totals[3]} beverages in the two grocery bags.")
```

## Teamwork

```python
ali = np.array([5, 4, 6, 4, 2])
ben = np.array([0, 5, 5, 3, 4])
carla = np.array([3, 6, 0, 3, 4])
```

```python
all_knocks = np.array([ali, ben, carla])
print(all_knocks)
```

```python
person_totals = np.sum(all_knocks, axis=1)
group_total = np.sum(all_knocks)

print(f"Ali: {person_totals[0]}, Ben: {person_totals[1]}, Carla: {person_totals[2]}")
print(f"Group total: {group_total}")
```

```python
daily_knocks = np.sum(all_knocks, axis=0)
rolling_knocks = np.cumsum(daily_knocks)
estimated_votes = rolling_knocks * 0.5

print(estimated_votes)
```
