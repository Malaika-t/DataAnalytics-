```python
import numpy as np
import pandas as pd
```

## Series

```python
aa = [1, 2, 3, 4]
ab = (5, 6, 7, 8)
ac = {9, 10, 11, 12}
ad = {'first': 13, 'second': 14, 'third': 15, 'fourth': 16}
```

```python
ser1 = pd.Series(aa)
ser2 = pd.Series(ab)
ser3 = pd.Series(list(ac))
ser4 = pd.Series(ad)

print("ser1:\n", ser1)
print("\nser2:\n", ser2)
print("\nser3:\n", ser3)
print("\nser4:\n", ser4)
```

```python
ser5 = pd.Series(17, 18, 19, 20)
print(ser5)

# Attempt to create ser5 with literal values 17, 18, 19, 20
# This raises a TypeError because pd.Series() expects a single iterable, can not pass multiple positional arguments.
# TypeError: Index(...) must be called with a collection of some kind, 18 was passed
```

## DataFrames from Series

```python
df1 = pd.DataFrame(ser1)
df1
```

```python
df1.insert(1, 'ser2', ser2)
df1.insert(2, 'ser3', ser3)
df1
```

```python
df1.insert(3, 'ser4', ser4)
df1
# ser4 was created from a dictionary (ad), so it has string index labels ('first', 'second', 'third', 'fourth') instead of integer positions (0-3).
# But df1 index is integer-based so pandas cannot align ser4 string labels to df1 integer index, resulting in NaN for all ser4 values.
```

```python
df1.loc[4] = [5, 9, 12, 0]
df1
```
