# Exercise 2.C Creating and calling functions
## Lab 1. working with defining custom functions

```python
def favorite_things(name, movie, music):
    return f"Name: {name} \nFavorite Movie: {movie} \nFavorite Musician/Band: {music}"

print(favorite_things("Malaika", "Inception", "Zayn Malik"))
```

```python
def why_im_here(purpose):
    return purpose

print(why_im_here("I joined the Year Up United Data Analyst program to build technical skills and build a career in data."))
```

```python
def favorite_place(place, why):
    return f"One of my favorite places is {place} because {why}."

print(favorite_place("Pakistan", "the food and culture are incredible"))
```

```python
print(favorite_place("California", "the natural beauty is unlike anywhere else in the world"))
```
