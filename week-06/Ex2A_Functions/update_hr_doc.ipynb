## Lab 3. Employee data 
This lab exercise makes updates to information extracted from an employee database.

```python
hr_list = [ 
    ('0123', 'HR', 'Rebecca Yang', '69000'),
    ('0121', 'IT', 'Mark Blick', '67000'), 
    ('0068', 'IT', 'Kahna Larsen', '112000'), 
    ('0234', 'OPS', 'Jim Smith', '54000') 
]
```

```python
# Update Mark's last name: 

hr_list[1] = (hr_list[1][0], hr_list[1][1], 'Mark Blick-Hawley', hr_list[1][3])
```

```python
print(hr_list)
```

```python
# Update Jim's department and salary:

hr_list[3] = (hr_list[3][0], 'CS', hr_list[3][2], '60000')
```

```python
print(hr_list)
```

```python
# Pipe formatting:

for emp in hr_list:
    print(f"{emp[0]} | {emp[1]} | {emp[2]} | {int(emp[3]):,}")
```
