# Exercise 2.B Using library functions
## Lab 2. Working with datetime functions

```python
import datetime
```

# Today

```python
today = datetime.datetime.now()
```

```python
print("Today is", today.strftime("%A, %B %d, %Y"))
```

# My birthday

```python
birthday = datetime.date(2000, 12, 19)
```

```python
print("My birthday is", birthday.strftime("%m/%d/%Y"))
```

# Next 90 Days

```python
ninety_d = today + datetime.timedelta(days=90)
```

```python
print("90 days from today is", ninety_d.strftime("%B %d, %Y"))
```

# Time for Dinner

```python
dinner_time = datetime.time(18, 30)
```

```python
print("Let's meet for dinner at", dinner_time.strftime("%H:%M PM"))
```
