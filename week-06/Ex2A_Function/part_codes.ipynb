# Exercise 2.A Using built-in functions 
## Lab 2. Parsing strings – part codes

```python
# format: supplier:productcode-size

# Functions:
def get_supplier_code(part_code):
    """Returns all characters before the colon."""
    return part_code[:part_code.index(':')]

def get_product_num(part_code):
    """Returns all characters between the colon and the dash."""
    colon_position = part_code.index(':')
    dash_position = part_code.index('-')
    return part_code[colon_position + 1:dash_position]

def get_size(part_code):
    """Returns all charcaters after the - """
    return part_code[part_code.index('-') + 1:]   
```

```python
# Create variables:

part1 = "ACME:123-L"
part2 = "DI:12345-M"
part3 = "ACE:1-12"
```

```python
# Call all three functions

for part_code in [part1, part2, part3]:
    supplier = get_supplier_code(part_code)
    product = get_product_num(part_code)
    size = get_size(part_code)
    print(f"For {part_code}, the supplier is {supplier}."
          f"Part numer #{product} is available in size {size}.")
```
